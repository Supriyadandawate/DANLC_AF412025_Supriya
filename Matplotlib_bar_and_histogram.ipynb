# Lab2: Create a line plot to visualize the daily closing prices of a stock over a year and give your conclusion.


```python
import matplotlib.pyplot as plt

# Given data
days = list(range(1, 78))
stock_prices = [
    100, 105, 110, 115, 112, 120, 118, 125, 128, 130, 132, 135,
    138, 140, 142, 144, 145, 148, 150, 155, 160, 158, 162, 165,
    170, 172, 175, 178, 180, 182, 185, 188, 190, 192, 195, 198,
    200, 198, 195, 193, 190, 188, 185, 182, 180, 178, 175, 172,
    170, 168, 165, 162, 160, 158, 155, 152, 150, 148, 145, 143,
    140, 138, 135, 132, 130, 128, 125, 123, 120, 118, 115, 112,
    110, 108, 105, 103, 100
]

# Plotting the stock prices
plt.figure(figsize=(12, 6))
plt.plot(days, stock_prices, marker='o', color='b', linestyle='-')
plt.title('Daily Closing Prices of a Stock Over a Year', fontsize=14)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Stock Price ($)', fontsize=12)
plt.grid(visible=True, linestyle='--', alpha=0.5)
plt.xticks(days[::5], rotation=45)  # Showing every 5th day for better readability
plt.tight_layout()
plt.show()
```

# Lab3: Create a bar chart to represent monthly expenses in different spending categories and give your conclusion.
Input:
categories = ['Rent', 'Groceries', 'Utilities', 'Entertainment', 'Transportation']
# Monthly expenses in dollars (replace with your own data)
expenses = [1200, 400, 200, 150, 250]

```python
import matplotlib.pyplot as plt

# Given data
categories = ['Rent', 'Groceries', 'Utilities', 'Entertainment', 'Transportation']
expenses = [1200, 400, 200, 150, 250]

# Creating a bar chart
plt.figure(figsize=(8, 5))
plt.bar(categories, expenses, color='skyblue')
plt.title('Monthly Expenses by Category', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Expense (in $)', fontsize=12)
plt.grid(visible=True, linestyle='--', alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

# Lab4: Create a histogram to represent the distribution of product prices in a retail store and give your conclusion.
Input:
product_prices = [24.99, 34.99, 49.99, 64.99, 39.99, 54.99, 79.99, 99.99, 29.99, 44.99,
59.99, 69.99, 84.99, 109.99, 119.99, 89.99, 74.99, 124.99, 69.99, 54.99]


```python
import matplotlib.pyplot as plt

# Given data
product_prices = [
    24.99, 34.99, 49.99, 64.99, 39.99, 54.99, 79.99, 99.99, 29.99, 44.99,
    59.99, 69.99, 84.99, 109.99, 119.99, 89.99, 74.99, 124.99, 69.99, 54.99
]

# Creating a histogram
plt.figure(figsize=(8, 5))
plt.hist(product_prices, bins=8, color='royalBlue', edgecolor='black')
plt.title('Distribution of Product Prices in a Retail Store', fontsize=14)
plt.xlabel('Price Range ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a dummy data frame
data = {
    'regions': ['North', 'South', 'East', 'West', 'Central'],
    'sales': [500, 300, 400, 350, 450]
}
df = pd.DataFrame(data)

# Calculating the percentage of sales for each region
total_sales = df['sales'].sum()
df['sales_percentage'] = (df['sales'] / total_sales) * 100

# Plotting the bar chart
plt.figure(figsize=(8, 5))
plt.bar(df['regions'], df['sales'], color='lightblue')
plt.title('Sales by Region', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.grid(visible=True, linestyle='--', alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

# Printing the data frame with sales percentages
print(df)
```
